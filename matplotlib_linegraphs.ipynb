# Subplots: multiple graphs on the same figure

```python
fig, ax = plt.subplots(1, 3)
fig.tight_layout() # This sometimes breaks formatting - check

ax[0].plot("r","1s", data=s_orbitals, linewidth = 2)
ax[1].plot("r","2s", data=s_orbitals, linewidth = 2)
ax[2].plot("r","3s", data=s_orbitals, linewidth = 2)

for axis in ax:
    axis.set_xlabel("r")
    axis.set_ylabel(r"$\psi$")
    axis.legend()
    axis.set_xlim(0, 14)
    axis.set_ylim(-0.05, 0.5)
```
